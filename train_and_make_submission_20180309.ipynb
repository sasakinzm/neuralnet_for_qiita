```python
import pandas as pd
df_sub = pd.read_csv("input/test.csv")
del df_sub["PassengerId"]
del df_sub["Name"]
del df_sub["Ticket"]
del df_sub["Cabin"]
df_sub["Sex"].replace("male", 0, inplace=True)
df_sub["Sex"].replace("female", 1, inplace=True)
df_sub["Embarked"].replace("Q", 0, inplace=True)
df_sub["Embarked"].replace("S", 1, inplace=True)
df_sub["Embarked"].replace("C", 2, inplace=True)
df_sub["Age"].fillna(29.7, inplace=True)
df_sub["Embarked"].fillna(1, inplace=True)
df_sub["Fare"].fillna(8.000, inplace=True)

test_array = np.array(df_sub)
```

```python
result_raw = train_neuralnet.network.predict(test_array)
result_raw
```

```python
results = []
passengerid_count = 892
for i in range(len(result_raw)):
    if result_raw[i][0] > result_raw[i][1]: results.append([passengerid_count, 0])
    if result_raw[i][0] <= result_raw[i][1]: results.append([passengerid_count, 1])
    passengerid_count += 1

results
```

```python
import csv
with open("result/predict_result_data.csv", "w") as f:
    writer = csv.writer(f, lineterminator='\n')
    writer.writerow(["PassengerId", "Survived"])
    for result in results:
        writer.writerow(result)
```
